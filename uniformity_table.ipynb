```python
def makeShare(numShare: int, origBoolExpr):
    varList = list(origBoolExpr.atoms(Symbol))
    numVars = len(varList)
    for i in range(numShare):
        strCommand = f"y{i+1} = symbols('y_{i+1}')"
        exec(strCommand)
        for j in range(numVars):
            strCommand = f"{varList[j]}_{i+1} = symbols('{varList[j]}_{i+1}')"
            exec(strCommand)
    globals().update(locals())
```

```python
makeShare(3, y)
```

```python
def findShare(num: int, x: bool):
    shareList = []
    for i in range(2**num):
        strList = list(bin(i)[2:].zfill(num))
        intList = list(map(int, strList))
        if(np.bitwise_xor.reduce(intList) == x):
            shareList.append(intList)
    return shareList
```

```python
findShare(3, 0)
```

```python
findShare(3, 1)
```

```python
def checkUniformity(num:int, origBoolExpr, shareExprList):
    varList     = list(origBoolExpr.atoms(Symbol))
    numVars     = len(varList)
    shareDict   = collections.OrderedDict()
    shareSubs   = [f'_{i+1}' for i in range(num)]
    shareVars   = [str(x)+y for x,y in list(itertools.product(*([varList]+[shareSubs])))]
    print(varList)
    for i in range(2**numVars):
        strList = list(bin(i)[2:].zfill(numVars))
        intList = list(map(int, strList)) # A Input Pattern
        print(intList,':')
        for j in range(numVars):
            shareDict[varList[j]] = findShare(num, intList[j])
        shareCombs = list(itertools.product(*shareDict.values()))
        for shareComb in shareCombs:
            shareValueList = list(itertools.chain.from_iterable(shareComb))
            print(shareValueList)
            for shareExpr in shareExprList:
                for idx, shareVar in enumerate(shareVars):
                    shareExpr = shareExpr.subs(shareVar, shareValueList[idx])
                    print(shareExpr)
                print(shareExpr%2)
    print(shareDict)
```
